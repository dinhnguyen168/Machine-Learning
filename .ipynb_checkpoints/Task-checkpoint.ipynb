### 1.2 Outliers:

```python
#indicate outliers

import matplotlib.pyplot as plt
import numpy as np

#calculate 2d indicators
def indicate(samples):
    #alternatively you can calulate any other indicators
    max = np.max(samples, axis=1)
    min = np.min(samples, axis=1)
    return max, min
```

```python
#Feature selection
# Ranking features
#paired t-tests Selector
from scipy.stats import ttest_ind

def pairedTTestRanker(data,label):
    featurePs=np.zeros(data.shape[1])
    for f in range(data.shape[1]):
        pVals=[]
        classes=np.unique(label)
        for i in range(len(classes)):
            for j in range(i,len(classes)):
                t, prob = ttest_ind(data[label==classes[i],f],data[label==classes[j],f])
                pVals.append(prob)
        featurePs[f]=np.mean(pVals)
    return np.argsort(featurePs)
```

There are many visual and statistical methods to detect outliers. In this report, we want to show two effective methods of finding outliers: **Histograms** and **Scatter Plot**

#### Scatter plot

Histograms identify values that are far away from the average values for each feature (univariate outliers). However, they fail to identify any abnormal behavior between two or more variables (multivariate outliers). That is when scatter plots come in handy.

A scatter plot represents the relationship between two numerical variables. The independent variable is plotted along the x-axis, while the dependent variable is plotted along the y-axis. A scatter plot helps us to visualize if two variables present a linear, polynomial, or another type of relationship, and how strong this relationship is.

Logically, we expect a relationship between the height of a woman and her weight. The following scatter plot shows the relationship between both variables.

```python
# indicate max and min
x,y = indicate(samples)
print(x[0],y[0])
plt.scatter(x, y, marker='x', c=labels, alpha=0.5, cmap='viridis')
plt.xlabel("Maximum of features of every sample")
plt.ylabel("Minimum of features of every sample")
plt.show()
```

```python
rank = pairedTTestRanker(samples, labels)
print(rank, rank[1])
plt.scatter(samples[:,rank[0]], samples[:,rank[1]], marker='x', c=labels, alpha=0.5, cmap='viridis')

plt.xlabel("First rank")
plt.ylabel("Second rank")
plt.show()
```

```python
f,ax = plt.subplots()

ax.scatter(samples[:,rank[0]],
           samples[:,rank[1]],
           c = labels,
           alpha=0.5,
           cmap = 'viridis')

# notation indicating a possible outlier
ax.annotate('Possible outlier',
            xy=(-990,-990),
            xytext=(-1000,-900),
            fontsize=12,
            arrowprops = dict(arrowstyle = '->',
                              ec = 'grey',
                              lw = 2),
            bbox = dict(boxstyle = "round",
                        fc="0.8"))

# ticks 
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)

plt.xlabel('1st ranked feature')
plt.ylabel('2rd ranked feature')
plt.show()
```

```python
# with outliers
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(xs = samples[:,rank[0]],
           ys = samples[:,rank[1]],
           zs = samples[:,rank[2]],
           zdir = 'z',
           s = 20, c = labels,
           cmap = 'viridis')

plt.title('In 3D representation')

plt.show()
```

#### Histograms:
The histogram we used here is for the purpose of visualizing the distribution of a numerical variable. In a histogram, the data is split into intervals also called bins. 

Each bars height represents the frequency of data points within each bin.
The histograms for both variables are shown below. The bars are displayed in the shape of a bell-shape curve which indicates that both features (weight and height) are normally distributed. Additionally, the Gaussian kernel density estimation function is depicted as well. This function is an approximation of the probability density function and represents the probability of a continuous variable to fall within a particular range of values. 

Since our data has a total of 12 features, we only take the 2 random features as the outliers representation

```python
# histogram and kernel density estimation function of the variable height
ax = sns.distplot(samples[:,rank[0]],
                  hist = True,
                  color = 'green',
                  hist_kws = {"edgecolor": 'w',
                              "linewidth": 10},
                  kde_kws = {"linewidth": 2})

# notation indicating a possible outlier
ax.annotate('Possible outlier',
            xy = (-990,0.0020),
            xytext = (-1000,0.0070),
            fontsize = 12,
            arrowprops = dict(arrowstyle = '->',
                              ec='grey',
                              lw=2),
            bbox = dict(boxstyle = "round",
                        fc = "0.8"))


# labels and title
plt.xlabel('1st ranked feature', fontsize = 14)
plt.ylabel('frequency', fontsize = 14)
plt.title('Distribution of 1st ranked feature', fontsize = 20);
```
